# Deep Learning Funda

* Deep learning is a cutting-edge machine learning technique that is based on representation learning--a powerful approach that allows machines to automatically learn high-level feature representations from data.

* Deep learning algorithms use an artificial neural network, a computing system which learns high-level features from data by increasing the depth (i.e., number of layers) in the network. The inspiration for neural networks comes partially from biological neural networks.

* A deep neural network is a specific type that excels at capturing nonlinear relationships in data.

A neural network is comprised of the following components:

1. Input Layer: This is where the training observations are fed through the independent variables.

2. Hidden Layers: These are the intermediate layers between the input and output layers. This is where the neural network learns about the relationships and interactions of the variables fed in the input layer.

3. Output Layer: This is the layer where the final output is extracted as a result of all the processing which takes place within the hidden layers.

4. Node: A node, also called a neuron, in a neural network is a computational unit that takes in one or more input values and produces an output value.



**Shallow NN**

* A shallow neural network is a neural network with a small number of layers--often one or two layers. Shallow neural networks are typically used for simple tasks, such as regression or classification.

* A simple shallow neural network with one hidden layer is shown below, where x1 and x2 are the two response variables feeding into the two nodes - n1 and n2 of the single hidden layer, which finally moves forward to generate the output.

image.png

**Deep NN**

* In contrast to the shallow neural network, a deep (dense) neural network is composed of multiple hidden layers, each of which contains a set of neurons that learn to extract certain features from the data. The output layer produces the final results of the network. The image below represents the basic architecture of deep neural networks with n-hidden layers.

* The extra hidden layers in a deep neural network allow it to learn more complex patterns than a shallow neural network. As a result, deep neural networks are more accurate than shallow neural networks, but they are also more computationally expensive to train. For this reason, shallow neural networks are sometimes used when computational resources are limited.

* For complex real time real world applications--such as multi-variate time series forecasting, natural language processing, real time forecasting, or predictive lead times--deep neural networks are preferable.

## Forward Propagation

In the basic neural network, you'll see that the neural network flows in the forward direction. This process of passing input data through the layer(s) of a neural network to generate an output is called forward propagation. The input data is first fed into the input layer, which passes it to the hidden layer(s). The hidden layer(s) process the data and pass it to the output layer, which produces the final output.

The calculations at the nodes are performed using the multiply then sum process - **dot product**; mostly an **activation function is used though to introduce non-linearity**.

image.png

```python
import numpy as np

input_layer = np.array([34, 7])
weights_topnode = np.array([2, -5])
weights_bottomnode = np.array([1, -2])
output_weights = np.array([3, -1])

top_node_value = np.dot(input_layer, weights_topnode).sum()
bottom_node_value = np.dot(input_layer, weights_bottomnode).sum()
print(top_node_value)    # 33
print(bottom_node_value) # 20
hidden_layer_values = np.array([top_node_value, bottom_node_value])
prediction = np.dot(hidden_layer_values, output_weights).sum()
print(prediction)        # 79
```

**Forward Propagation in Deep Neural Networks**

image.png

```python
input_layer = np.array([34, 7])
weights_node_n1 = np.array([2, -5])
weights_node_n2 = np.array([1, -2])
weights_node_n3 = np.array([3, -2])
weights_node_n4 = np.array([1, -1])
output_weights = np.array([2, -6])

n1 = np.dot(input_layer,weights_node_n1).sum()
n2 = np.dot(input_layer,weights_node_n2).sum()
layer_1_output = np.array([n1,n2])
print(layer_1_output)  # [33, 20]

n3 = np.dot(layer_1_output,weights_node_n3).sum()
n4 = np.dot(layer_1_output,weights_node_n4).sum()
layer_2_output = np.array([n3,n4])
print(layer_2_output)  # [59, 13]

prediction = np.dot(layer_2_output,output_weights).sum()
print(prediction)      # 40
```

## Activation Functions

* The multiply then sum calculation is a linear operation, and activation functions are used to introduce non-linearity into the output of a neuron.

* **Sigmoid Activation Function** has horizontal asymptotes at 0 and 1, which means any input value will always output a value between 0 and 1. The sigmoid activation is an ideal activation function for a binary classification problem where the output is interpreted as a binomial probability distribution.

image.png

* **Softmax Activation Function** are probabilities that range from 0 to 1. The sum of all the probabilities should equal 1, as each value represents the probability of a certain class label being the correct one. For example, if an input is classified into 3 classes, then the softmax output will be three values between 0 and 1 where they add up to 1. While sigmoid is used for binary classification problems, the softmax activation function is better for multiclass classification tasks.

* **ReLU Activation Function** is far more computationally efficient compared to the sigmoid and tanh functions. ReLU stands for rectified linear unit and is defined as follows: `ReLU(x) = max(0, x)`

* **Tanh Activation Function**  takes any real value and outputs a value between -1 and +1. The larger your input value, the closer the output will get to 1; whereas the smaller the input value, the closer the output will be to -1. Below, we can see a plot for tanh with an input range of [-20, 20]:

image.png

**Implementation of ReLU in a deep nn**

image.png

```python
input_layer = np.array([34, 7])
weights_node_n1 = np.array([2, -5])
weights_node_n2 = np.array([1, -5])
weights_node_n3 = np.array([3, -2])
weights_node_n4 = np.array([1, -1])
output_weights = np.array([2, -4])

def relu(input_value):
    output_value = max(0, input_value)
    return(output_value)

n1 = np.dot(input_layer,weights_node_n1).sum()
n2 = np.dot(input_layer,weights_node_n2).sum()
n1_output = relu(n1)
n2_output = relu(n2)
print(n1_output)      # 33
print(n2_output)      # 0

layer_1_output = np.array([n1_output,n2_output])
print(layer_1_output) # [33, 0]

n3 = np.dot(layer_1_output,weights_node_n3).sum()
n4 = np.dot(layer_1_output,weights_node_n4).sum()
n3_output = relu(n3)
n4_output = relu(n4)
print(n3_output)      # 99
print(n4_output)      # 33

layer_2_output = np.array([n3_output, n4_output])
print(layer_2_output) # [99, 33]

prediction = np.dot(layer_2_output,output_weights).sum()
print(prediction)     # 66
```

## Multi Layer Modeling & Optimizers

```python
target_variable = ['mpg']
predictors = list(set(list(df.columns)) - set(target_variable))
df[predictors] = df[predictors] / df[predictors].max()

X = df[predictors].values
y = df[target_variable].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=100)

print(X_train.shape); print(X_test.shape)


model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(32, input_shape=(X_train.shape[1],), activation='relu'))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(8, activation='relu'))
model.add(tf.keras.layers.Dense(4, activation='relu'))
model.add(tf.keras.layers.Dense(1))

optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(loss='mae', metrics=['mae'], optimizer=optimizer)

model.fit(X_train, y_train, epochs=150, verbose=0)
print(model.evaluate(X_train, y_train))
print(model.evaluate(X_test, y_test))
```

* Previously, we learned how to build a multi-layer deep neural network with the ReLU activation function. We achieved an R-squared value of 0.92 and 0.86 on train and test data, respectively. Note that you may get a slightly different result than this, which is fine! The objective is to learn the process and enjoy the journey!

* Now, we'll build the multi-layer neural network with the ELU activation function, but before we discuss that, let's revisit the ReLU function quickly.

* We know that the ReLU function is defined as follows: `ReLU(x)=max(0,x)`

* The max(0, x) function call returns the maximum value between 0 and x, which means that:

 * when x is less than or equal to 0, the value 0 is returned, and
 * when x is greater than 0, the value x is returned

* The limitation of this function is in cases where we have several negative input values, for example, daily stock price changes. In such cases, ReLU will take these negative values and will return a value of zero. This is undesirable because we know stock price changes can be negative, so we'll want to capture this in our activation function.

* In situations like these, we can use the ELU activation function, which stands for exponential linear unit. The function has a hyperparameter alpha that controls the value when input values are negative. This is explained by the ELU function below:

* The ELU function is defined with alpha > 0 as:

 * when x is less than or equal to 0, the value alpha * (exp(x) - 1) is returned, and
 * when x is greater than 0, the value x is returned

 image.png

* Note that the alpha value is always greater than zero. It takes the default value of one. As a refresher, the exp(x) function returns e^x, where
e is the base of the natural logarithm constant. This is also known as the exponential function.

```python
# elu as activation function in Dense Layer (instead of relu)

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(32, input_shape=(X_train.shape[1],), activation='elu'))
model.add(tf.keras.layers.Dense(16, activation='elu'))
model.add(tf.keras.layers.Dense(8, activation='elu'))
model.add(tf.keras.layers.Dense(4, activation='elu'))
model.add(tf.keras.layers.Dense(1))

optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(loss='mae', metrics=['mae'], optimizer=optimizer)


model.fit(X_train, y_train, epochs=150, verbose=0)
print(model.evaluate(X_train, y_train))
print(model.evaluate(X_test, y_test))

pred_train = model.predict(X_train)
print(r2_score(y_train, pred_train))

pred_test = model.predict(X_test)
print(r2_score(y_test, pred_test))
```

* We achieved the R-squared value of:

 * 0.92 on train data and 0.86 on test data with the ReLU activation function.

 * 0.91 on train data and 0.87 on test data with the ELU activation function.

* Both activation functions give similar performance on train and test data.

* Apart from the activation functions, another important factor in the deep neural network architecture is the optimizer. Until now, we've been using the Adam optimizer, but there are other optimizers we can use. One such optimizer is the stochastic gradient descent optimizer, represented by the SGD class in TensorFlow's Keras.

* SGD works by iteratively moving through the weights and updating them based on how well they are performing. SGD optimizes a function by iteratively moving in the direction of steepest descent, as defined by the negative gradient of the function. In simple terms, the goal is to find the set of weights that results in the lowest error rate.

* Keras allows us to specify many different parameters for SGD, such as the `learning_rate`, `momentum`, etc.

* The learning rate is a parameter that controls how much the weights of the network are updated at each training step.

* `Momentum` is another parameter that accelerates the gradient descent in the suitable direction. The momentum parameter takes values greater than or equal to zero, and the default value is zero. You can access the SGD class through `tf.keras.optimizers.SGD()`.

```python
# using SGD optimizer

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(32, input_shape=(X_train.shape[1],), activation='relu'))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(8, activation='relu'))
model.add(tf.keras.layers.Dense(4, activation='relu'))
model.add(tf.keras.layers.Dense(1))

optimizer = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.5)
model.compile(loss='mae', metrics=['mae'], optimizer=optimizer)

model.fit(X_train, y_train, epochs=150, verbose=0)
print(model.evaluate(X_train, y_train))
print(model.evaluate(X_test, y_test))

pred_train = model.predict(X_train)
print(r2_score(y_train, pred_train))

pred_test = model.predict(X_test)
print(r2_score(y_test, pred_test))

plt.scatter(y_test, pred_test)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.show()
```

* We achieved the R-squared value of:

 * 0.92 on train data and 0.86 on test data with the ReLU activation function and Adam optimizer.

 * 0.91 on train data and 0.87 on test data with the ELU activation function and Adam optimizer.

 * 0.85 on train data and 0.81 on test data with the ELU activation function and SGD optimizer.

* The above comparison shows that the last model--with the ELU activation function and SGD optimizer--underperformed on the train and test data.

* We now have learned how to use the SGD optimizer to build a multi-layer deep neural network. One more popular optimizer choice is Root Mean Square Propagation, represented by the RMSprop class in the `tf.keras.optimizers` module.

* It works by keeping track of the moving average of the squares of the gradients and dividing this by the square root of the sum of all previous squared gradients. This results in faster learning for models with many training examples.

* Apart from `learning_rate`, another important argument in the RMSprop optimizer is `rho`, which is a **discounting factor that is applied to the history/coming gradient**. A higher `rho` value means that the learning rate will decay more slowly, which can help the optimization process converge on a better solution. However, too high of a rho value can also lead to problems such as overfitting.

* The best value for any given problem will depend on the specifics of the data and model being used. It has a **default value of 0.9.**

```python
# using RMSProp Optimizer

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(32, input_shape=(X_train.shape[1],), activation='relu'))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(8, activation='relu'))
model.add(tf.keras.layers.Dense(4, activation='relu'))
model.add(tf.keras.layers.Dense(1))

optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.1, rho=0.8)
model.compile(loss='mae', metrics=['mae'], optimizer=optimizer)


model.fit(X_train, y_train, epochs=150, verbose=0)
print(model.evaluate(X_train, y_train))
print(model.evaluate(X_test, y_test))

pred_train = model.predict(X_train)
print(r2_score(y_train, pred_train))

pred_test = model.predict(X_test)
print(r2_score(y_test, pred_test))

plt.scatter(y_test, pred_test)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.show()
```

* We achieved the R-squared value of:

 * 0.92 on train data and 0.86 on test data with the ReLU activation function and Adam optimizer.

 * 0.91 on train data and 0.87 on test data with the ELU activation function and Adam optimizer.

 * 0.85 on train data and 0.81 on test data with the ELU activation function and SGD optimizer.

 * 0.83 on train data and 0.75 on test data with the ReLU activation function and RMSprop optimizer.

* The above comparison shows that the last two models underperformed significantly, compared to the first two models, on the train and test data.

* In all the neural networks we've built so far, we only used one type of activation function when adding a new layer to the model. However, it's possible to use a different activation function for each hidden layer in a multi-layer deep neural network.

```python
# model with different configurations
  ## nodes in dense layers
  ## different activation functions for layers = relu & elu
  ## Adam optimizer

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(500, input_shape=(X_train.shape[1],), activation='relu'))
model.add(tf.keras.layers.Dense(300, activation='relu'))
model.add(tf.keras.layers.Dense(100, activation='elu'))
model.add(tf.keras.layers.Dense(20, activation='elu'))
model.add(tf.keras.layers.Dense(4, activation='elu'))
model.add(tf.keras.layers.Dense(1))

optimizer = tf.keras.optimizers.Adam(learning_rate=0.1)
model.compile(loss='mae', metrics=['mae'], optimizer=optimizer)


model.fit(X_train, y_train, epochs=150, verbose=0)
print(model.evaluate(X_train, y_train))
print(model.evaluate(X_test, y_test))

pred_train = model.predict(X_train)
print(r2_score(y_train, pred_train))

pred_test = model.predict(X_test)
print(r2_score(y_test, pred_test))
```

We achieved an R-squared value of:

* 0.92 on train data and 0.86 on test data with the ReLU activation function and Adam optimizer.

* 0.91 on train data and 0.87 on test data with the ELU activation function and Adam optimizer.

* 0.85 on train data and 0.81 on test data with the ELU activation function and SGD optimizer.

* 0.83 on train data and 0.75 on test data with the ReLU activation function and RMSprop optimizer.

* 0.85 on train data and 0.82 on test data by changing node configurations and using a mix of ReLU and ELU activation functions at various layers, with Adam optimizer.

* There are five categorical such variables- `embarked_town`, `sex`, `adult_male`, `alone`, and `who`-that require dummy encoding.

* Although the variable `pclass` currently contains integer values, they are actually "**integer labels**," and we'll need to treat them as if they were categorical values as well.

### Modeling & Evaluation

Unlike the Sequential model, where the input shape is defined in the first hidden layer itself, the Functional model defines a standalone input layer that specifies the shape of input data. This is shown below:

```python
input_layer = tf.keras.Input(shape=(X_train.shape[1],))
```

The next step is defining the hidden layers, which from the Functional API perspective, are comprised of the following:

* specifying the number of nodes
* selecting an activation function
* connecting the incoming layer

```python
hidden_layer1 = tf.keras.layers.Dense(128, activation='relu')(input_layer)
```

**Dropout Layers**

The idea behind a dropout layer is to randomly drop units (either hidden or visible) from the network during training. This has the effect of making the network more resistant to overfitting.

image.png

```python
# dropout layer with a dropout rate = 0.4
drop1 = tf.keras.layers.Dropout(rate=0.40)(hidden_layer1)
```

```python
# Further layers
hidden_layer2 = tf.keras.layers.Dense(64, activation='relu')(drop1)
drop2 = tf.keras.layers.Dropout(rate=0.2)(hidden_layer2)
hidden_layer3 = tf.keras.layers.Dense(16, activation='relu')(drop2)
hidden_layer4 = tf.keras.layers.Dense(8, activation='relu')(hidden_layer3)
hidden_layer5 = tf.keras.layers.Dense(4,activation='relu')(hidden_layer4)
```

The last hidden layer, `output_layer`, of the model is used as the output layer. The last layer is shown below, which is a dense layer with a `sigmoid` activation function.

```python
output_layer = tf.keras.layers.Dense(1, activation='sigmoid')(hidden_layer5)
```

```python
# Model instantiation
model = tf.keras.Model(inputs=input_layer, outputs=output_layer)
```

```python
# Compiling the model
optimizer = tf.keras.optimizers.Adam(0.001)
loss = tf.keras.losses.BinaryCrossentropy()
metrics = ['accuracy']
model.compile(optimizer=optimizer,loss=loss, metrics=metrics)
```

```python
# Fitting the model (Model Training)
model.fit(X_train, y_train, epochs=150, verbose=0)
```

```python
# Model Evaluation on train data
model.evaluate(X_train, y_train)
```

```python
# Model Evaluation on test data
model.evaluate(X_test, y_test)
```
